## ANALYSIS 

### 1.What is the average age of our customers?

```python
# Calculating the mean age of the customers 
Avg_age = retail['Age'].mean().round(0)
f"The average age of the customers is {Avg_age} years"
```

### 2. What is the gender distribution of our customers?

```python
# Plotting a histogram to see the distribution of gender 
```

```python
(retail.Gender.value_counts().plot(kind = 'bar',
                                   y = 'Number of customers',
                                   title = 'Gender Distribution',
                                   ylabel = 'Number of customers',color = 'cyan'));
```

#### There are more Male customers than Female customers. 

### 3.What is the most common customer segment?

```python
# Finding the segment with the highest number of customers
Most_common_Cust_Seg = retail["Customer_Segment"].value_counts().sort_values(ascending = False)
```

```python
Most_common_Cust_Seg
```

```python
#plotting a pie chart for customer segmentation
Most_common_Cust_Seg.plot(kind = 'pie');
```

#### The most common segment is the Regular followed by the New segment

### 4. What is the total amount of sales made in 2023?

```python
# using calculating the total sales for 2023
Total_Amount_Sales_2023 = retail[retail["Year"]== 2023]['Total_Amount'].sum().round(2)
```

```python
Total_Amount_Sales_2023
print(f"The total Sales amount made in 2023 is ${Total_Amount_Sales_2023}")
```

### 5. What is the average purchase amount per customer?

```python
#using group by to calculate average purchase amount
Avg_purchase_amnt_customer = retail.groupby("Customer_ID")["Total_Purchases"].mean().round(0)
print(Avg_purchase_amnt_customer )
```

### 6. Which product category has the highest sales?

```python
#Grouping by product category to find the highest sales amount
Highest_Sales_Product_Category = retail.groupby("Product_Category")['Total_Amount'].sum().sort_values(ascending = True)
```

```python
# plotting a horizontal bar chart to present the findings
Highest_Sales_Product_Category.plot.barh( color = 'magenta', title = 'Total Sales By Product Category',  xlabel= 'Total Sales');
```

#### The electronics category has the highest amount of sales 

### 7. What are the top 5 most popular products?

```python
# finding  the most demanded  products 
Top_5_most_pop_products = retail.products.value_counts().sort_values(ascending = False).head(5)
```

```python
Top_5_most_pop_products
```

### 8. Which brand has the highest sales?

```python
#sorting product brands by the total sales
Brand_highest_sales = retail.groupby('Product_Brand')['Total_Amount'].sum().sort_values(ascending = False)
```

```python
#plotting a brand vs total sales bar chart
Brand_highest_sales.plot(kind = 'bar', title = 'Brand by Total Sales', ylabel = 'Total sales', color = 'b', figsize= (12,8));
```

#### The brand with the highest sales is Pepsi

### 9. What is the most common shipping method used?

```python
#counting transations by each shipping method
Most_common_shipping_method = retail.Shipping_Method.value_counts().
```

```python
#plotting outcome
Most_common_shipping_method.plot(kind = 'bar', title = 'The most common shipping method', ylabel = 'Count of transactions ' );
```

### 10. What is the most common payment method used?

```python
# calculating the average of ratings 
Average_rating = retail.Ratings.mean().round(2)
```

### 12. How many customers have given excellent feedback?

```python
#calculating the total number of excellent feedback
Excellent_Feedback = retail[retail["Feedback"]== 'Good'].value_counts().sum()
```
